```python


subjects = ['Maths', 'Physics', 'Chemistry', 'English']

df['Total'] = df[subjects].sum(axis=1)

(df[['Student_ID', 'Student_Name', 'Total']].head())
```

```python
df['average'] = df[['Maths', 'Physics', 'Chemistry', 'English']].values.mean()

df[['Student_Name', 'average']].tail(7)
```

```python

df['Average'] = df[['Maths', 'Physics', 'Chemistry', 'English']].mean(axis=1)

df['Grade'] = 'D'


df.loc[df['Average'] >= 60, 'Grade'] = 'C'
df.loc[df['Average'] >= 70, 'Grade'] = 'B'
df.loc[df['Average'] >= 80, 'Grade'] = 'A'
df.loc[df['Average'] >= 90, 'Grade'] = 'A+'
df[['Student_Name', 'Average', 'Grade']].tail(5)
```

```python
top_10 = df.nlargest(10, 'Total')
("Top 10 Students")
(top_10[['Student_ID', 'Student_Name', 'Total']])
```

```python
df['Rank'] = df['Total'].rank(method='min', ascending=False)
("\nRank-wise Analysis")
(df[['Student_Name', 'Total', 'Rank']].sort_values('Rank').head(10))

```

```python
weak_students = df[
    (df['Maths'] < 40) |
    (df['Physics'] < 40) |
    (df['Chemistry'] < 40) |
    (df['English'] < 40)
]

("\nStudents Scoring Below 40")
(weak_students[['Student_ID', 'Student_Name']])
```

```python
import pandas as pd
import matplotlib.pyplot as plt


df['Average'] = df[['Maths', 'Physics', 'Chemistry', 'English']].mean(axis=1)


dept_avg = df.groupby('Department')['Average'].mean()

dept_avg.plot(kind='bar',
              color=['purple', 'blue', 'red', 'pink', 'green'])

plt.title('Average Marks by Department')
plt.xlabel('Department')
plt.ylabel('Average Marks')

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt



dept_count = df['Department'].value_counts()

dept_count.plot(
    kind='pie',
    autopct='%1.1f%%'
)

plt.title('Department-wise Student Distribution')

plt.ylabel('')
plt.show()
```
